```python
deposit = list(df2['Deposits'])
deposit_list = []
```

```python
for i in range(len(deposit)):
    if isinstance(deposit[i], int):
        deposit_list.append(deposit[i])
    else:
        new = deposit[i].replace('$', "", 1)
        sub = "."
        new2 = new[:new.index(sub) + len(sub)]
        new2 = new2.replace('.', '', 1)
        new2 = new2.replace(',', '', 1)
        deposit_list.append(int(new2))
        
```
